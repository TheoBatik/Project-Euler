## Problem 17 Number letter counts
https://projecteuler.net/problem=17

```python
#Dictionaries for reference: based on https://stackoverflow.com/questions/15598083/python-convert-numbers-to-words
ones = {"1":"One","2":"Two","3":"Three","4":"Four","5":"Five","6":"Six", "7":"Seven","8":"Eight","9":"Nine"}
afterones = {"10":"Ten","11":"Eleven","12":"Twelve","13":"Thirteen","14":"Fourteen","15":"Fifteen","16":"Sixteen", "17":"Seventeen","18":"Eighteen","19":"Nineteen"}
tens = {"2":"Twenty","3":"Thirty","4":"Forty","5":"Fifty","6":"Sixty", "7":"Seventy","8":"Eighty","9":"Ninety"}
grand={0:" Billion, ",1:" Million, ",2:" Thousand, ", 3:""}
```

```python
# Function that converts numbers to words up to 1000
def num2word(num):
    num = str(num)
    length = len(num)
    word = ''
    if length > 4:
        return('Number is too long to be converted.')
    if length == 1:
        word += ones[num] 
    if length == 2:
        if num[0] == '1':
            word += afterones[num]
        else:
            word += tens[num[0]]
            if num[1] != '0':
                word += '-' + ones[num[1]]      
    if length == 3:
        word += ones[num[0]] + ' hundred'
        if num[1] == '1':
            word += ' and ' + afterones[num[1:]]
        else:
            if num[1] != '0':
                word += ' and ' + tens[num[1]]
            if num[2] != '0':
                word += '-' + ones[num[2]] 
    if length == 4:
        word = 'One Thousand'
    return(word)

#  Function that counts the number of letters in a string
def letter_count(word):
    count = 0
    for letter in word:
        print('letter', letter)
        if letter != ' ' and letter != '-':
            count += 1
        print('count', count)
    return(count)        
```

```python
# test functions
n = 45
print(num2word(n))
print(letter_count(num2word(n)))
```

```python
count = 0
start = 1 #1
end = 1000
for n in range(start, end + 1):
    print(num2word(n))
    count += letter_count(num2word(n))
    print(count)
print(count)  
```
